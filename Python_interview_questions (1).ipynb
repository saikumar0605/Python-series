## 32. Create a custom context manager without using contextlib?

````python
class CustomFileManager:
  def __init__(self, filename, mode):
    self.filename = filename
    self.mode = mode
    self.f = None

  def __enter__(self):
    self.f = open(self.filename, self.mode)
    return self.f

  def __exit__(self, exc_type, exc_value, exc_traceback):
    self.f.close()
    
````

````python
with CustomFileManager("ex.txt", "w") as f:
  f.write("Test")
````

````python
class Email:
  def __init__(self, name):
    self.name = name

  def __enter__(self):
    self.email = self.name + "@gmail.com"
    return self.email

  def __exit__(self, exc_type, exc_value, exc_traceback):
    print("your email ID is created")
````

````python

class LengthError(Exception):
  def __init__(self, message):
    super().__init__(message)

def mean_squared_error(l1, l2):
  try:
    if not (isinstance(l1,list) and isinstance(l2, list)):
      raise TypeError("Type of Argument should be list")
    if len(l1)!=len(l2):
      raise LengthError("Length of Both List should match") # LengthError
    error = [] 
    for el1, el2 in zip(l1,l2):
      error.append((el1-el2)**2)

    return sum(error)/len(error)

  except Exception  as e :
    raise e
````

````python
def minmax(*args):
  for ele in args:
    if not isinstance(ele, (int, float)):
      raise TypeError("args must be of type int or float")
  return min(args), max(args)
````

## 49. Write a function to create transpose of matrix? without using numpy or any third party library?

````python
def transpose(matrix):
  if not isinstance(matrix, list):
    raise TypeError("Type must be a list")
  try:
    N = len(matrix[0])
  except Exception as e:
    raise e

  for i in range(len(matrix)):
    if not isinstance(matrix[i], list):
      raise TypeError("matrix rows should be a list")
    if N != len(matrix[i]):
      raise Exception("Shape error")

  row_N, col_N = len(matrix), len(matrix[0])
  result = [[0]*row_N for _ in range(col_N)] # zero matrix as a placeholder
  for row in range(row_N):
    for col in range(col_N):
      result[col][row] = matrix[row][col]

  return result
````

````python
'''
Author: Abe Alexander
'''

class InvalidVector(Exception):
    def __init__(self, message):
        super().__init__(message)

class Vector():
    def __init__(self, *args):
        self.args = args
        
    def __add__(self, Vector2):
        if len(self.args) != len(Vector2.args):
            raise InvalidVector("Vector Length does not match")
        return list(map(sum, zip(self.args, Vector2.args)))
        
    def __sub__(self, Vector2):
        if len(self.args) != len(Vector2.args):
            raise InvalidVector("Vector Length does not match")
        return list(map(lambda x: x[0]-x[1], zip(self.args, Vector2.args)))
````

````python
class Vector:
  """
  This is a Vector class
  It takes int or float values
  """
  def __init__(self, *args):
    if len(args) == 0:
      raise ValueError
    for arg in args:
      if not isinstance(arg, (int, float)):
        raise TypeError

    self.vec  = args

  def __str__(self):
    return f"{self.vec}"

  def __repr__(self):
    return f"Vector class object: Vector({self.vec})"

  def __len__(self):
    return len(self.vec)

  def __abs__(self):
    result = (sum(element**2 for element in self.vec))**(1/2)
    return result

  def __add__(self, other):
    add = tuple(a+b for a, b in zip(self, other))
    return self.__class__(*add)

  def __sub__(self, other):
    add = tuple(a-b for a, b in zip(self, other))
    return self.__class__(*add)

  def __iter__(self):
    return self.vec.__iter__()

  def __getitem__(self, key):
    return self.vec[key]
````
